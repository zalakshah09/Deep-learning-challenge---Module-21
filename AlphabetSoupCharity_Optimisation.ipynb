## Preprocessing

```python
# Import our dependencies
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf

#  Import and read the charity_data.csv.

charity_df = pd.read_csv("Resources/charity_data.csv")
charity_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
target = ["IS_SUCCESSFUL"]
features = charity_df.drop(columns=["EIN", "IS_SUCCESSFUL"])
```

```python
# Determine the number of unique values in each column.
unique_counts = features.nunique()

# Print the unique value counts for each column
print(unique_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_counts = features["APPLICATION_TYPE"].value_counts()
app_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE

application_types_to_replace = list(app_counts[app_counts < 500].index)


# Replace in dataframe
for app in application_types_to_replace:
    charity_df['APPLICATION_TYPE'] =  charity_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
charity_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
class_counts = features["CLASSIFICATION"].value_counts()

# Print top 10 classification categories
class_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
class_counts =features["CLASSIFICATION"].value_counts()
class_counts_greater_than_one = class_counts[class_counts >1]
class_counts_greater_than_one
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Set the cutoff value and create list of classifications to be replaced
cutoff = 1000
class_counts = charity_df["CLASSIFICATION"].value_counts()
classifications_to_replace = list(class_counts[class_counts < cutoff].index)

# Replace in dataframe
for cls in classifications_to_replace:
    charity_df['CLASSIFICATION'] = charity_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
charity_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
encoded_features= pd.get_dummies(features)
encoded_features.head()
```

```python
# Split our preprocessed data into our features and target arrays

# target = "IS_SUCCESSFUL"
# features = application_df.drop(["EIN", "NAME", target], axis=1)
# Split the preprocessed data into a training and testing dataset
X = encoded_features
y = charity_df[target].values

# Split the data into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

#Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
from tensorflow.keras.models import Sequential
nn =Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=7, input_dim=len(X_train.columns), activation='relu'))
nn.add(tf.keras.layers.Dense(units=14, input_dim=len(X_train.columns), activation='sigmoid'))
nn.add(tf.keras.layers.Dense(units=21, input_dim=len(X_train.columns), activation='sigmoid'))



# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
import os
from tensorflow.keras.callbacks import ModelCheckpoint

checkpoint_path = "AlphabetSoupCharity_Optimisation.h5"

# Create a callback that saves the model's weights every five epochs
cp_callback = ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=2,
    save_weights_only=True,
    save_freq=5*len(X_train_scaled))
```

```python
# Train the model
# nn.fit(X_train_scaled, y_train, epochs=100, batch_size=32, callbacks=[cp_callback])
nn.fit(X_train_scaled, y_train, epochs=100, shuffle=True, verbose=2)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity.h5")
```
